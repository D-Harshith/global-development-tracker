```python
import pandas as pd
from sqlalchemy import create_engine
import os
engine = create_engine(os.getenv('DATABASE_URL', 'postgresql://postgres:your_password@localhost:5432/global_development_tracker'))
```

```python
co2 = pd.read_csv('co2-data.csv')
co2 = co2[['country', 'year', 'iso_code', 'population', 'co2', 'co2_per_capita']]
co2 = co2.astype(str)
co2 = co2.replace('nan', None)
print(co2.head(2))

co2.to_sql('co2', engine, schema='staging', if_exists='append', index=False)
print(f"co2 loaded: {len(co2)} rows")
```

```python
life = pd.read_csv("life-expectancy.csv")
life.columns= ['entity', 'code', 'year', 'life_expectancy']
life = life.astype(str)
life = life.replace('nan', None)
print(life.head(2))

life.to_sql('life_expectancy', engine, schema='staging', if_exists='append', index=False)
print(f"life loaded: {len(life)} rows")
```

```python
gdp = pd.read_csv("GDP_per_capita.csv")
gdp = gdp.drop(columns=['Indicator Name', 'Indicator Code'])
year_columns = [col for col in gdp.columns if str(col).strip().isdigit()]

gdp_long = gdp.melt(
    id_vars=['Country Name', 'Country Code'],
    value_vars = year_columns,
    var_name='year',
    value_name = 'value'
)

gdp_long.columns=['country_name','country_code','year','value']
gdp_long = gdp_long.astype(str)
gdp_long = gdp_long.replace('nan', None)

gdp_long.to_sql('gdp_per_capita', engine, schema='staging', if_exists='append', index=False)
print(f"GDP loaded: {len(gdp_long)} rows")
```

```python
pop = pd.read_csv('Population.csv')
# pop.columns
pop = pop.drop(columns=['Indicator Name', 'Indicator Code'])
year_column = [col for col in pop.columns if str(col).strip().isdigit()]

pop_shot = pop.melt(
    id_vars =['Country Name', 'Country Code'],
    value_vars = year_column,
    var_name='year',
    value_name = 'value'
)
pop_shot.columns = ['country_name', 'country_code', 'year','value']
pop_shot = pop_shot.astype(str)
pop_shot = pop_shot.replace('nan', None)

pop_shot.to_sql('population', engine, schema='staging', if_exists='append', index=False)
```
